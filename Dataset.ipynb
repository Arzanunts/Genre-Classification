```python
gtzan_genre = dict(list(zip(genresgt,[0 for _ in range(10)])))
for genre in genresgt:
    ans = np.zeros((171,))
    songs = os.listdir(path2+f'/{genre}')
    for j,s in tqdm(enumerate(songs)):
        if j == minsong-1:
            break
        if s == '.DS_Store':
            pass
        dur = pydub.utils.mediainfo(path2+f'/{genre}'+f'/{s}')['duration']
        data, sr = lr.load(path2+f'/{genre}'+f'/{s}',  duration = math.floor(float(dur)))
        data = data[:617400]
        n_fft = int(sr * 0.02)  # Длина окна в отсчетах (20 мс при sr=22050 Гц)
        hop_length = int(n_fft / 2)  # Шаг перекрытия в отсчетах (50% перекрытие)
        n_mels = 128  # Количество мел-фильтров
        S = lr.feature.melspectrogram(y=data, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        chroma_cqt = np.mean(lr.feature.chroma_cqt(y=data,sr=sr),axis=1)
        m_htk = np.mean(lr.feature.mfcc(y=data, sr=sr, dct_type=3),axis=1)
        m_htk = np.append(m_htk,np.mean(S,axis=1))
        m_htk = np.append(m_htk,(chroma_cqt))
        m_htk = np.append(m_htk,
                          (
          np.mean(lr.feature.spectral_centroid(y=data, sr=sr)),
          np.mean(lr.feature.spectral_bandwidth(y=data, sr=sr)),
          np.mean(lr.feature.spectral_contrast(y=data,sr=sr)),
          np.mean(lr.feature.spectral_flatness(y=data)),
          np.mean(lr.feature.spectral_rolloff(y=data,sr=sr))
        )
                         )
        ans = np.vstack((m_htk,ans))
    gtzan_genre[genre] = ans
```
